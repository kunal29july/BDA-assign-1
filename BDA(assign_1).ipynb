```
Y_array = np.asarray(data[['retweets']]) #converting column of retweets into list/array
```

```
#printing retweets column into array
y_arr_final=[]
for i in Y_array:
  i=i[0]
  y_arr_final.append(i)
y_arr_final.sort()
print(y_arr_final)
```

```
#dictionary which show "retweets" and "frequency" where key is retweets and value is frequency
d={}
for i in y_arr_final:
  if i not in d:
    d[i]=1
  else:
    d[i]=d[i]+1
print(d)
```

```
#frequency array:
fr_arr=[]
for i in d:
  fr_arr.append(d[i])
print(fr_arr)

```

```
#sum of frequency array except, frequency of "0" and "1"
for i in range(2,len(fr_arr)):
  sum=sum+fr_arr[i]
print(sum)
```

```
#sum of frequencies:
#print(d)
sum=0
for i in d:
  sum=sum+d[i]
sum_of_freq=sum
print(sum_of_freq)
```

```
#cummulative of frequencies
sum=0
cumm_freq=[]
for i in d:
  sum=sum+d[i]
  cumm_freq.append(sum)
print(cumm_freq)
```

```
#cummulative dictionary:
j=0
for i in d:
  print("retweets:" ,i   ,  "frequency:", d[i]   ,"cummulative freq:" ,cumm_freq[j])
  j=j+1
```

**Finding quartiles based on formula shown below in program**

```
#first quartiles:
N1=sum_of_freq
q_1=1*(N1+1)/4
print(q_1)#6423.75 th item
```

```
#spliting  item
res1=[]
for i in cumm_freq:
  if (i>q_1):
    res1.append(i)

    
qq1=min(res1)
print(min(res1))
  
```

**retweets: 0 frequency: 15827 cummulative freq: 15827**

```
#third quantile:
N3=sum_of_freq
q_3=3*(N3+1)/4
print(q_3)#6423.75 th item
```

```
#spliting  item
res3=[]
for i in cumm_freq:
  if (i>q_3):
    res3.append(i)
    

qq3=min(res3)    
print(min(res3))
  
```
